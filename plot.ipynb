```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Version 1 CIFARS10 Models/training_loss.csv',header=None,names=['loss'])
df_1 = pd.read_csv('./Version 1 CIFARS10 Models/training_loss1shot.csv',header=None,names=['loss'])
df_5 = pd.read_csv('./Version 1 CIFARS10 Models/training_loss5shot.csv',header=None,names=['loss'])
df_10 = pd.read_csv('./Version 1 CIFARS10 Models/training_loss10shot.csv',header=None,names=['loss'])
```

```python
rolling_avg = df['loss'].rolling(window=1400).mean()
rolling_avg1shot = df_1['loss'].rolling(window=1400).mean()
rolling_avg5shot = df_5['loss'].rolling(window=1400).mean()
rolling_avg10shot = df_10['loss'].rolling(window=1400).mean()
```

```python
plt.plot(df['loss'], label='Training loss',linewidth=0.05)
plt.plot(rolling_avg, color='red', linewidth=1, label='Moving average')
plt.title('Training Loss of FSL Method, 50,000 Episodes')
```

```python
plt.plot(df_1['loss'],linewidth=0.05,color='b')
plt.plot(df_5['loss'],linewidth=0.05,color='g')
plt.plot(df_10['loss'],linewidth=0.05,color='purple')
plt.plot(rolling_avg1shot,color='red',linewidth=1, label='1-shot')
plt.plot(rolling_avg5shot,color='yellow',linewidth=1, label='5-shot')
plt.plot(rolling_avg10shot,color='darkorange',linewidth=1,label='10-shot')
plt.legend()
```
